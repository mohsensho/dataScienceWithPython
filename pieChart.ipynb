```python
# import required libraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

# job data in percentile
job_data = ['40', '20', '17', '8', '5', '10']
# define Label as different departments
labels = 'IT', 'Finance', 'Marketing', 'Admin', 'HR', 'Operations'

# explode the 1st and 2nd slices which is IT, Finance
explode = (0.04,0.03,0,0,0,0)
# draw the pie chart and set the parameters
plt.pie (job_data,labels=labels, explode=explode)
# show the plot
plt. show()
```

```python
# demo 2 with showing percentage and start angle of the bigger portion
#  job data in percentile
job_data = ['40', '20', '17', '8', '5', '10']
# define Label as different departments
labels = 'IT', 'Finance', 'Marketing', 'Admin', 'HR', 'Operations'


#set the pie chart plot properties
#set figure size
plt.figure(figsize=(4,4))
#explode the largest pie in the data set
explode = (0.05,0,0,0,0,0)
#set pie chart properties with autopct for showing percentage and start angle of the bigger portion at 45 degrees
plt.pie(job_data,labels=labels, explode=explode,autopct="%1.1f%%", startangle=45)
#set axis equal to draw pie as circle
plt.axis('equal')
#set title of the pie chart
plt.title('Ohio State - 2012 : Leading Causes of Death')
#show the plot (pie chart)
plt. show
```
