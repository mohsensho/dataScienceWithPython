```python
#import the boston dataset from sklearn Library
from sklearn.datasets import load_diabetes
#import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
#Load boston dataset
diabetes_dataset = load_diabetes()
```

```python
#view boston dataset
print(diabetes_dataset.DESCR)
```

```python
#define x axis for the data
x_axis = diabetes_dataset.data
x_axis
```

```python
#define y axis for the target
y_axis = diabetes_dataset.target
y_axis
```

```python
#plot histogram
style.use('classic')
plt.figure(figsize=(7,7))
plt.hist(y_axis, bins=50)
plt.xlabel('Data')
plt.ylabel('Target')
plt.show()
```

```python
#available style for the plot
style.available
```

```python
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.scatter(diabetes_dataset.data[:,5], diabetes_dataset.target)
plt.xlabel('Data')
plt.ylabel('Target')
plt.show()
```
