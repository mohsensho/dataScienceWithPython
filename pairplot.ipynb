```python
#import the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#To view the plot in notebook
%matplotlib inline
```

```python
#import the auto data set
df_auto_dataset = pd.read_csv('Datasets/OSL Datasets/auto_data_mpg.csv')
```

```python
# view the top 5 records
df_auto_dataset.head()
```

```python
#write a user defined function for origin
# 1-USA, 2-Europe, 3-Asia
def origin(num):
    if num==1: 
        return 'USA'
    elif num==2:
        return 'Europe'
    else:
        return 'Asia'
#use apply function to chane origin code to continent name
df_auto_dataset['origin'] = df_auto_dataset['origin'].apply(origin)
```

```python
#view first 30 data points after applying the user defined function to data set
df_auto_dataset.head(30)
```

```python
# draw the pair plot using sns for mpg, weight, orign and with hue origin, set the size to 4
# note: hue is variable in data set to map plot aspects to different colors
# all options for kind: ‘scatter’, ‘kde’, ‘hist’, ‘reg’
sns.pairplot(df_auto_dataset[['mpg','weight', 'origin']], hue='origin', height=4, kind="scatter")
```

```python
# set diag_kind to kde (the default is auto) other options are ‘auto’, ‘hist’, ‘kde’, None
sns.pairplot(df_auto_dataset[['mpg','weight', 'origin']], hue='origin', height=4, diag_kind="hist")
```
